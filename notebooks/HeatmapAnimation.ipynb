```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib inline
plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 100  
plt.ioff()

# Initialize data
data = np.random.rand(3, 3)

# Set up the plot
fig, ax = plt.subplots()
heatmap = sns.heatmap(data, annot=True, fmt=".2f", cmap="viridis", cbar=True, ax=ax)

# Customize tick labels
ax.set_xticks(range(3))
ax.set_yticks(range(3))
ax.set_xticklabels(['a', 'b', 'c'])
ax.set_yticklabels(['a', 'b', 'c'])

# Update function for animation
def update(frame):
    print(type(frame), frame)
    ax.clear()
    # Update the heatmap with new data
    new_data = np.random.rand(3, 3)  # Replace with your actual data source
    heatmap = sns.heatmap(new_data, annot=True, fmt=".2f", cmap="viridis", cbar=False, ax=ax)
    ax.set_xticks(range(3))
    ax.set_yticks(range(3))
    ax.set_xticklabels(['a', 'b', 'c'])
    ax.set_yticklabels(['a', 'b', 'c'])
    return [heatmap]

# Create animation
ani = FuncAnimation(fig, update, frames=10, interval=500, blit=False)
ani
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 100  
plt.ioff()

# Initialize plot
fig, ax = plt.subplots()
t = np.linspace(0, 3, 40)
g = -9.81
v0 = 12
z = g * t**2 / 2 + v0 * t

v02 = 5
z2 = g * t**2 / 2 + v02 * t

# Create initial plot elements
scat = ax.scatter(t[0], z[0], c="b", s=5, label=f'v0 = {v0} m/s')
line2 = ax.plot(t[0], z2[0], label=f'v0 = {v02} m/s')[0]
ax.set(xlim=[0, 3], ylim=[-4, 10], xlabel='Time [s]', ylabel='Z [m]')
ax.legend()

# Update function for animation
def update(frame):
    x = t[:frame]
    y = z[:frame]
    # Update scatter plot
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # Update line plot
    line2.set_xdata(t[:frame])
    line2.set_ydata(z2[:frame])
    return (scat, line2)

# Create animation
anim = animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)

# animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)
anim
```
